```python
# Comenzare el proyecto cargando mis librerias necesarias

import pandas as pd
import sidetable as stb
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
# Cargo mi dataset
game_data = pd.read_csv("C:/Users/audre/OneDrive/Escritorio/tripletel/dataset_p6/games.csv")
```

```python
# Ahora leere los datos basicos de mi dataset: 

def EDA (data):
    print(data.head())
    print()
    print(data.info())
    print()
    print('Filas duplicadas\n', data.duplicated().sum())
    print()
    print('Numero de datos nulos\n', data.isna().sum())

    
EDA(game_data)

# Mis observaciones
#Podemos ver que Las columnas con valores nulos son: Name, Year_of_Release, Genre, Critic_Score, User_Score, y Rating
# Hay que cambiar el formato de las columnas year_of_release y user_score segun un primer analisis
```

```python
# Utilizaremos sidetable para checar el numero de duplicados que tenemos en nuestras columnas y mas

game_data.stb.missing(style=True)
```

```python
# Como primer paso de limpieza de datos ambiaremos los encabezados a minusculas
game_data.columns = game_data.columns.str.lower()
display(game_data.head())
```

```python
# Hacemos labores de limpieza de datos: 
game_data['year_of_release'] = game_data['year_of_release'].fillna(0).astype(int)
game_data['name']=game_data['name'].fillna('unknown')
game_data['genre']=game_data['genre'].fillna('unknown') 
game_data['rating']=game_data['rating'].fillna('unknown') 

# Cambiamos el tipo de datos de la columna user_score la cual tiene valores tipo str 
game_data.loc[game_data['user_score']== 'tbd','user_score']=np.nan
game_data['user_score']=game_data['user_score'].astype(float)
game_data.info()
```

```python
# Como primer analisis crearemos con una grafica hacia donde van nuestros valores atipicos
sns.boxplot(game_data['user_score'])
plt.show()
```

```python
# Ya que la columna user_score ya ha sido convertida a float, procederemos a rellenar los valores faltantes con la media de los datos.
# (Este paso solo se realiza para temas de buen manejo de datos)

game_data['user_score'] = game_data['user_score'].fillna(game_data['user_score'].median())
print(game_data.info())  #confirmamos
```

```python
# Mismo pprocedimiento para critic_score, podemos ver los valores atipicos abajo
sns.boxplot(game_data['critic_score'])
plt.show()
```

```python
game_data['critic_score'] = game_data['critic_score'].fillna(game_data['critic_score'].median())
print(game_data.info()) #confirmamos
```

```python
# Ya que las callificaciones en la columna rating son a traves de letras, esta columna la rellenaremos con el comentario unknown
game_data['rating'] = game_data['rating'].fillna('unknown')
game_data.info()
```

```python
# Ya que estan limpios nuestros datos con las columnas correctas y sin datos nulos vamos a continuar 

# Agregamos una columna con nuestros datos de venta totales

game_data['total_sales']=game_data[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)

# Quiero esta columna a un lado de las ventas
total_sales_col = game_data.pop('total_sales')
game_data.insert(8, 'total_sales', total_sales_col)
game_data.head()
```

```python
# Ahora analizemos cuantos juegos se lanzan por a√±o y quitamos los  que tienen un 0 en el a√±o

games_by_yr=game_data.groupby('year_of_release').count()['name'].sort_values().reset_index()
games_by_yr = games_by_yr.query('year_of_release != 0') # para mi: recuerda el signo de ! para diferente de 0
games_by_yr.columns=['year_of_release','total_games']
games_by_yr.head(10)
```

```python
# Comenzamos a hacer un buen grafico para los juegos sacados por a√±o
fig, ax = plt.subplots(figsize=(11, 6))

ax.vlines(x=games_by_yr.year_of_release, ymin=0, ymax=games_by_yr.total_games, color='purple',linewidth=10)
ax.set_title('Games Released Annually')
ax.set_xlabel('Year of release')
ax.set_ylabel('Number of games released')
ax.set_xticks(games_by_yr['year_of_release'].unique())
plt.xticks(rotation=60)

plt.show()

# Mis observaciones

# Se observa un crecimiento constante en las decadas de 1980 y 1990, con un aumento significativo a partir del a√±o 2000 y un punto maximo entre 2008 y 2009 donde se lanzaron mas de 1400 juegos. 
# Sin embargo a partir de 2010 se nota una disminucion en la cantidad de lanzamientos
```

```python
# Ahora haremos un grafico para las ventas de cada plataforma pero comenzamos agrupando nuestros datos principales

games_sales=game_data.groupby('platform')['total_sales'].sum().sort_values(ascending=False).reset_index()
display(games_sales.head(10))
```

```python
plt.figure(figsize=(15, 6)) 
total_sales_graf = sns.barplot(data=games_sales, x="platform", y="total_sales", color='purple')
total_sales_graf.set(title='Most popular platforms', xlabel='platform', ylabel='Total Sales')
plt.xticks(rotation=60) 
plt.show()

# Mis observaciones:
# Podemos ver que PS2 cuenta con el primer lugar de ventas con mas de 1,200
```

```python
# Obtendremos el zscore que es la distancia de la que  las ventas totales estan para con el average

games_sales['z_score']=(games_sales['total_sales']-games_sales['total_sales'].mean())/games_sales['total_sales'].std()
games_sales.head(10)
```

```python
games_sales['color']=['red' if puntaje < 0 else 'green' for puntaje in games_sales['z_score']]
```

```python
#Graficamos como en clase y podemos ver que los valores en negativo son valores que estan por debajo de la media, mientras que los valores positivos son valores por encima de la media

plt.figure(figsize=(15,10))
plt.hlines(
    y=games_sales.platform, xmax=games_sales.z_score,xmin=0,linewidth=10,alpha=0.5, color=games_sales.color
)
plt.show()

# Mis observaciones
# Podemos ver que las mejores plataformas son las que estan por encima de la media de las ventas totales :)
```

```python
# Ahora crearemos nuestra grafica donde podremos ver el historico de ventas por juego cada a√±o
# Comenzamos con una tabla pivot solo del a√±o 2011 en adelante como recomendacion del a√±o

cleaned_data=game_data[game_data.year_of_release>=2011]
selected_games = ['PS2','X360', 'PS3', 'Wii','DS'] # las plataformas con mas ventas segun nuestra grafica anterior
full_table = cleaned_data.query("platform in @selected_games")
pivot=cleaned_data.pivot_table(index='year_of_release',columns='platform',values='total_sales',aggfunc='sum').fillna(0)
pivot


sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(data=pivot)
plt.title('Sells per platform')
plt.xlabel('Year of release')
plt.ylabel('Total Sells')
plt.show()

# Podemos observar que las plataformas tienden a durar 10 a√±os 
```

```python
# Procederemos a probar las siguientes hipotesis:

#Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

xbo = cleaned_data.loc[cleaned_data.platform == 'XOne', 'user_score'].dropna()
pc = cleaned_data.loc[cleaned_data.platform == 'PC', 'user_score'].dropna()

#Utilizaremos  la pruebla de levene como se vio en discord
alpha = 0.05
levene_test = st.levene(xbo, pc)
print(f'p-valor de Levene: {levene_test.pvalue}')

if levene_test.pvalue < alpha:
    t_test = st.ttest_ind(xbo, pc, equal_var=False)  #utilizamos twelch o tstudent segun comentario, recuerda true para el segundo
else:
    t_test = st.ttest_ind(xbo, pc, equal_var=True)

print(f"La p es de: {t_test.pvalue}")

if t_test.pvalue < alpha:
    print("Los puntajes si son diferentes entre usuarios de PC y Xbox One")
else:
    print("Los puntajes son iguales")
```

```python
# Graficamos la igualdad:
hip1 = cleaned_data[cleaned_data.platform.isin(['PC', 'XOne'])]
hip1_graf = sns.barplot(data=hip1, x="platform", y="user_score", hue="platform")
hip1_graf.set(title='XOne vs PC in Score', xlabel="Platform", ylabel='user score')
hip1_graf.set_ylim(0, 8)
plt.show()
```

```python
# Procedemos a la siguiente hipotesis:
# Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes

# Comenzamos como antes
accion = cleaned_data.loc[cleaned_data.genre == 'Action', 'user_score'].dropna()
deportes = cleaned_data.loc[cleaned_data.genre == 'Sports', 'user_score'].dropna()

alpha = 0.05
levene_test = st.levene(accion, deportes)
print(f'p-valor de Levene: {levene_test.pvalue}')

if levene_test.pvalue < alpha:
    t_test = st.ttest_ind(accion,deportes, equal_var=False)   #de arriba
else:
    t_test = st.ttest_ind(accion,deportes, equal_var=True)

print(f'La p es de: {t_test.pvalue}')

if t_test.pvalue < alpha:
    print('Los puntajes si son diferentes entre los generos de Accion y Deportes')
else:
    print('Los puntajes son iguales')
```
